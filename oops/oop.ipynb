## Real-world analogy 1:

**I want to reuse the same function for Team Members and Managers, but they have small differences. How do I do that?**

- Both Team Members and Managers should be able to log in, but:

    - Team Members → should log in, view tasks, and submit tasks.

    - Managers → should log in, view tasks, and assign tasks.

```python
# defining a function:

def login(user_data):
    print(f"{user_data ['name']} has logged in ")

def view_task(user_data):
    print(f"{user_data ['name']} is viewing the task ")

def submit_task(user_data):
    if user_data['role'] =='team_member':
        print(f"{user_data['name']} has submitted the task")
    else :
        print ('Manager no need to submit the tasks')

def assign_task(user_data):
    if user_data['role'] =='manager':
        print(f"{user_data['name']} has assign the task")
    else :
        print ('team members unable to assign the task')

# data Collection
user_data = {'name' : 'Allan', 'role' : 'manager'}

#function to be called :
login(user_data)
view_task(user_data)
submit_task(user_data)
assign_task(user_data)
```
